```python
import numpy as np
```

```python
drought = np.loadtxt("1982_0.25x0.25_degree.txt")
flood = np.loadtxt("1983_0.25x0.25_degree.txt")
normal = np.loadtxt("1984_0.25x0.25_degree.txt")

drought = drought/np.sum(drought)
flood = flood/np.sum(flood)
normal = normal/np.sum(normal)
```

```python
normal = normal[normal>0]
drought = drought[drought>0]
flood = flood[flood>0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
norm_hist, norm_bins = np.histogram(normal, bins=np.arange(np.min(normal), np.max(normal), step=0.0001), density=True)
drut_hist, drut_bins = np.histogram(drought, bins=np.arange(np.min(drought), np.max(drought), step=0.0001), density=True)
```

```python
fig, ax = plt.subplots(3,1, figsize=(15,8))
fig.tight_layout()

ax[0].hist(normal, bins=np.arange(np.min(normal), np.max(normal), step=0.0001))
ax[0].set_yscale('log')
ax[0].set_title('Normal Monsoon')

ax[1].hist(flood, bins=np.arange(np.min(flood), np.max(flood), step=0.0001))
ax[1].set_yscale('log')
ax[1].set_title('Excess Monsoon')

ax[2].hist(drought, bins=np.arange(np.min(drought), np.max(drought), step=0.0001))
ax[2].set_yscale('log')
ax[2].set_title('Drought Monsoon')

plt.show()
```
